### Cleaning up data

```python
import pandas as pd
import numpy as np

df = pd.read_csv("assets/zone_prd.csv", skiprows=1, low_memory=False)

df = df.rename(columns={
    'Wa_num' : 'well_id',
    "Compltn_event_seq" : "comp_sq_id",
    "Prod_period": "date_raw",
    "UWI": "UWI",
    "Area_code" : "area_code",
    "Formtn_code": "frmtn_code",
    "Pool_seq":"pool_seq",
    "Gas_prod_vol (e3m3)": "gas_prod_e3m3",
    "Oil_prod_vol (m3)": "oil_prod_m3",
    "Water_prod_vol (m3)" : "water_prod_m3",
    "Cond_prod_vol (m3)": "cond_prod_m3",
    "Prod_days": "prod_days",
    "Project_code" : "project_code"
})

df['formation_name'] = df['frmtn_code'].map(formation_map).fillna(df['frmtn_code'])
df['date'] = pd.to_datetime(df['date_raw'], format='%Y%m', errors="coerce")
df = df.dropna(subset=['well_id', 'date'])


df['total_boe_m3'] = (
    df['oil_prod_m3'].fillna(0) + 
    df['cond_prod_m3'].fillna(0) + 
    df['gas_prod_e3m3'].fillna(0)
)

df['total_boe_bbl'] = df['total_boe_m3'] * 6.2898

print(f"Cleaned Data: {df.shape[0]} rows")

df.head(5)
```

### Plot Prod Function


```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

def plot_frmtn_professional(df, formation_code, start_date='2010-01-01', formation_name=None):
  
    if formation_name is None:
        try:
            names = df.loc[df['frmtn_code'] == formation_code, 'formation_name'].dropna().unique()
            formation_name = names[0] if len(names) > 0 else formation_map.get(formation_code, str(formation_code))
        except Exception:
            formation_name = formation_map.get(formation_code, str(formation_code))


    df_formation = df[df['frmtn_code'] == formation_code].copy()
    
 
    prod_cols = ['gas_prod_e3m3', 'cond_prod_m3', 'oil_prod_m3']
    
    monthly = df_formation.groupby('date')[prod_cols].sum()
    monthly = monthly[monthly.index >= start_date]

    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True, 
                                   gridspec_kw={'height_ratios': [3, 1]})
    
  
    ax1.fill_between(monthly.index, monthly['gas_prod_e3m3'] * 0.00003531, 
                    color='#003f5c', alpha=0.8, label='Natural Gas (Bcf)')

   
    ax1_twin = ax1.twinx()
    
    ax1_twin.plot(monthly.index, (monthly['cond_prod_m3'] + monthly['oil_prod_m3']) * 6.2898 / 1_000_000, 
                color='#b97800', linewidth=2, linestyle='--', label='Total Liquids (MMbbl)')

    
    ax1.set_title(f"{formation_name} Production Profile", fontsize=16, pad=15)
    ax1.set_ylabel("Gas Production (Bcf)", color='#003f5c', fontweight='bold')
    ax1_twin.set_ylabel("Liquids Production (MMbbl)", color="#b97800", fontweight='bold', rotation=270, labelpad=20)

   
    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax1_twin.get_legend_handles_labels()
    ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

 
    try:
       
        covid_date = pd.to_datetime('2020-03-01')
        if covid_date in monthly.index:
            val = monthly.loc[covid_date, 'gas_prod_e3m3'] * 0.00003531  # Convert to Bcf
            ax1.annotate('COVID-19 Shock', xy=(covid_date, val), xytext=(covid_date, val*1.2),
                        arrowprops=dict(facecolor='black', shrink=0.05))
    except:
        pass  

  
    cgr = (monthly['cond_prod_m3'] * 6.2898) / (monthly['gas_prod_e3m3'] * 0.03531)

    ax2.plot(monthly.index, cgr, color='#bc5090', linewidth=2)
    ax2.fill_between(monthly.index, cgr, color='#bc5090', alpha=0.1)

    ax2.set_ylabel("CGR (bbl/MMcf)", fontweight='bold')
    ax2.set_title("Liquid Yield (Condensate-Gas Ratio)", fontsize=12)
    ax2.set_xlabel("Year")


    ax2.xaxis.set_major_locator(mdates.YearLocator())
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

    footer_text = f"Data: BC Energy Regulator | Analysis: Shahbaz S.Mann"

    plt.figtext(0.99, 0.01, footer_text, 
                horizontalalignment='right', 
                fontsize=10, 
                style='italic', 
                color='#555555',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=5))

    plt.tight_layout()
    plt.show()
```

### Analysis of Well Drilling Activity

All top formations except Montney have experienced production declines in recent years. The data reveals a significant shift in drilling activity concentrated in the Montney formation.


```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def plot_activity_comparison(df, formation_code, formation_name=None, author_name="Your Name"):
   
    if formation_name is None:
        try:
            names = df.loc[df['frmtn_code'] == formation_code, 'formation_name'].dropna().unique()
            formation_name = names[0] if len(names) > 0 else formation_map.get(formation_code, str(formation_code))
        except Exception:
            formation_name = formation_map.get(formation_code, str(formation_code))

    
    formation_counts = df.groupby('frmtn_code')['UWI'].nunique().sort_values(ascending=False)
    top_5_codes = formation_counts.head(5).index.tolist()
    
   
    top_20 = formation_counts.head(20)

  
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
    
   
    names_display = [formation_map.get(x, str(x)) for x in top_20.index]
    y_pos = list(range(len(top_20)))
    bar_colors = ['#ffa600' if x == formation_code else '#b0b0b0' for x in top_20.index]

    ax1.barh(y_pos, top_20.values, color=bar_colors)
    ax1.set_yticks(y_pos)
    ax1.set_yticklabels(names_display)
    ax1.invert_yaxis() 
    
    ax1.set_title(" Top 20 Formations (by Total Wells)", fontsize=14, pad=15)
    ax1.set_xlabel("Total Unique Wells")
    ax1.grid(axis='x', alpha=0.3)
    
    montney_count = top_20[formation_code]
    montney_idx = list(top_20.index).index(formation_code)
    ax1.text(montney_count, montney_idx, f" {montney_count:,}", va='center', fontweight='bold', color='#ffa600')

  
    

    for code in top_5_codes:
      
        formation_df = df[df['frmtn_code'] == code]
    
        active_wells = formation_df.groupby('date')['UWI'].nunique()
        active_wells = active_wells[active_wells.index >= '2010-01-01']
        

        if code == formation_code:
    
            ax2.plot(active_wells.index, active_wells.values, 
                     color='#ffa600', linewidth=3, label=f"{formation_name} (Active)", zorder=10)
         
            ax2.fill_between(active_wells.index, active_wells.values, color='#ffa600', alpha=0.1)
        else:
         
            label_name = formation_map.get(code, f"Fmtn {code}")
            ax2.plot(active_wells.index, active_wells.values, 
                     linewidth=1.5, alpha=0.7, label=f"{label_name}")

    ax2.set_title(f"Active Well Stock: {formation_name} vs. Top Competitors", fontsize=14, pad=15)
    ax2.set_ylabel("Count of Producing Wells")
    ax2.grid(True, alpha=0.3)
    ax2.legend(loc='upper left')
    

    ax2.xaxis.set_major_locator(mdates.YearLocator(2))
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

 
    plt.figtext(0.99, 0.01, f"Data: BC Energy Regulator | Analysis: {author_name}", 
                horizontalalignment='right', 
                fontsize=10, style='italic', color='#555555',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=5))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.1) 
    plt.show()


plot_activity_comparison(df, 5000, formation_name=None, author_name="Shahbaz S.Mann")
```

### Pool-Level Development Analysis

Montney wells continue to be drilled extensively, significantly surpassing well counts in other formations. The following analysis examines the specific pools receiving the most development activity.


```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

def plot_pool_dynamics(df, formation_code, author_name="Your Name"):
   
    try:
        names = df.loc[df['frmtn_code'] == formation_code, 'formation_name'].dropna().unique()
        formation_name = names[0] if len(names) > 0 else formation_map.get(formation_code, str(formation_code))
    except Exception:
        formation_name = formation_map.get(formation_code, str(formation_code))

   
    df_fmtn = df[df['frmtn_code'] == formation_code].copy()
    
  
    top_pools = df_fmtn.groupby('pool_seq')['UWI'].nunique().sort_values(ascending=False).head(5).index.tolist()

    fig, ax = plt.subplots(figsize=(14, 8))
    
   
    colors = sns.color_palette("husl", n_colors=len(top_pools))
    

    for i, pool_id in enumerate(top_pools):
      
        pool_data = df_fmtn[df_fmtn['pool_seq'] == pool_id]
        
     
        activity = pool_data.groupby('date')['UWI'].nunique().cumsum()
        
       
        activity = activity[activity.index >= '2010-01-01']
        
       
        if i == 0:
            ax.plot(activity.index, activity.values, 
                label=f"Pool {pool_id} ", 
                color='#003f5c', linewidth=3, zorder=10)
            ax.fill_between(activity.index, activity.values, color='#003f5c', alpha=0.05)
        elif i == 1:
            ax.plot(activity.index, activity.values, 
                label=f"Pool {pool_id}", 
                color=colors[i], linewidth=2, alpha=0.8)
        else:
            ax.plot(activity.index, activity.values, 
                label="Others" if i == 2 else "",  # Only label once for legend
                color=colors[i], linewidth=2, alpha=0.5, linestyle='--')

    # 5. Styling
    ax.set_title(f"Sub-Play Development: {formation_name} Pool Activity Evolution", fontsize=16, pad=20)
    ax.set_ylabel("Active Producing Wells (Log Scale)", fontweight='bold')
    ax.set_xlabel("Year", fontweight='bold')
    
  
    ax.set_yscale('log')
    
   
    ax.grid(True, which="both", ls="-", alpha=0.2)
    
   
    ax.legend(title="Reservoir Pool", loc="upper left", bbox_to_anchor=(1, 1))
    
    
    ax.xaxis.set_major_locator(mdates.YearLocator(2))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

    plt.figtext(0.99, 0.01, f"Data: BC Energy Regulator | Analysis: {author_name}", 
                horizontalalignment='right', fontsize=10, style='italic', color='#555555',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=5))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.1, right=0.85) 
    plt.show()


plot_pool_dynamics(df, 5000, author_name="Shahbaz S.Mann")
```

## Making the Function

```python


from scipy.optimize import curve_fit


def dca_generator_by_well(montney_df, UWI, qabd, years):
    """
    DCA with SINGLE PLOT.
    Legend now shows: Model Name | EUR | Remaining Life
    """
    
    # ─────────────────────────────────────────────────────
    # STEP A: DATA PREPARATION & UNIT FIX
    # ─────────────────────────────────────────────────────
    
    # 1. Filter Data
    plot_data = montney_df[montney_df['UWI'] == UWI].copy()
    plot_data['date'] = pd.to_datetime(plot_data['date'])
    plot_data = plot_data.sort_values('date')
    
    # 2. Filter Time Window
    cutoff_date = pd.Timestamp.now() - pd.DateOffset(years=years)
    plot_data = plot_data[plot_data['date'] >= cutoff_date]
    
    if plot_data.empty:
        print(f"❌ ERROR: No data for well {UWI} in last {years} years")
        return

    # 3. UNIT CONVERSION
    if 'gas_prod_e3m3' in plot_data.columns:
        plot_data['gas_prod_m3'] = plot_data['gas_prod_e3m3'] * 1000
    elif 'gas_prod_m3' not in plot_data.columns:
        print("❌ ERROR: Could not find 'gas_prod_e3m3' or 'gas_prod_m3' column.")
        return

    # 4. Calculate Daily Rates
    plot_data['days_online'] = (plot_data['date'] - plot_data['date'].min()).dt.days
    plot_data['days_in_period'] = plot_data['days_online'].diff().fillna(30.44)
    plot_data['rate_daily'] = plot_data['gas_prod_m3'] / plot_data['days_in_period']
    
    x_data = plot_data['days_online']
    y_data = plot_data['rate_daily'] 
    cum_prod = plot_data['gas_prod_m3'].sum()
    last_day = x_data.max()

    # ─────────────────────────────────────────────────────
    # STEP B: CURVE FITTING
    # ─────────────────────────────────────────────────────
    qi_guess = y_data.max()
    bounds_hyp = (0, [np.inf, np.inf, 2.0]) # b-factor max 2.0
    
    try:
        popt_exp, _ = curve_fit(exponential, x_data, y_data, p0=[qi_guess, 0.005])
        popt_hyp, _ = curve_fit(hyperbolic, x_data, y_data, p0=[qi_guess, 0.005, 0.5], bounds=bounds_hyp)
        popt_harm, _ = curve_fit(harmonic, x_data, y_data, p0=[qi_guess, 0.005])
    except RuntimeError as e:
        print(f"❌ Fit Failed: {e}")
        return

    # ─────────────────────────────────────────────────────
    # STEP C: METRICS & PLOTTING SETUP
    # ─────────────────────────────────────────────────────
    # Find abandonment times
    t_abd_exp = find_t_abd(exponential, popt_exp, qabd, last_day)
    t_abd_hyp = find_t_abd(hyperbolic, popt_hyp, qabd, last_day)
    t_abd_harm = find_t_abd(harmonic, popt_harm, qabd, last_day)
    
    # Calculate Reserves (Returns: Remaining, EUR, Life_Years)
    metrics_exp = get_metrics(last_day, t_abd_exp, exponential, popt_exp, cum_prod)
    metrics_hyp = get_metrics(last_day, t_abd_hyp, hyperbolic, popt_hyp, cum_prod)
    metrics_harm = get_metrics(last_day, t_abd_harm, harmonic, popt_harm, cum_prod)

    # Setup Plot Axis
    valid_abds = [t for t in [t_abd_exp, t_abd_hyp, t_abd_harm] if t is not None]
    # Limit plot to max 25 years (9125 days) to avoid infinite harmonic lines squishing the view
    max_plot_days = min(max(valid_abds)*1.1 if valid_abds else last_day*1.5, 9125)
    
    t_smooth = np.linspace(x_data.min(), max_plot_days, 200)
    date_smooth = [plot_data['date'].min() + pd.Timedelta(days=float(x)) for x in t_smooth]

    # Create ONE figure
    fig, ax = plt.subplots(figsize=(12, 7), facecolor='white')
    
    # 1. Plot History (Black Dots)
    ax.plot(plot_data['date'], y_data, 'o', color='black', alpha=0.3, markersize=4, label='Historical Data')

    # 2. Plot Models with "Life" in Legend
    # Exponential
    # metrics[2] is the Life in Years
    life_exp = metrics_exp[2]
    lbl_exp = f"Exponential (Life: {life_exp:.1f} yrs | EUR: {metrics_exp[1]/1e6:.1f} E6m³)"
    y_exp = exponential(t_smooth, *popt_exp)
    ax.plot(date_smooth, y_exp, color='#D32F2F', linestyle='--', linewidth=2, label=lbl_exp)

    # Hyperbolic (Usually the best fit, make it solid)
    life_hyp = metrics_hyp[2]
    lbl_hyp = f"Hyperbolic (Life: {life_hyp:.1f} yrs | EUR: {metrics_hyp[1]/1e6:.1f} E6m³)"
    y_hyp = hyperbolic(t_smooth, *popt_hyp)
    ax.plot(date_smooth, y_hyp, color='#388E3C', linewidth=3, label=lbl_hyp)

    # Harmonic
    life_harm = metrics_harm[2]
    lbl_harm = f"Harmonic (Life: {life_harm:.1f} yrs | EUR: {metrics_harm[1]/1e6:.1f} E6m³)"
    y_harm = harmonic(t_smooth, *popt_harm)
    ax.plot(date_smooth, y_harm, color='#1976D2', linestyle=':', linewidth=2, label=lbl_harm)

    # 3. Plot Abandonment Line
    ax.axhline(qabd, color='orange', linestyle='-', linewidth=1.5, label=f'Economic Limit ({qabd:.0f} m³/d)')

    # ─────────────────────────────────────────────────────
    # STEP D: FORMATTING
    # ─────────────────────────────────────────────────────
    from matplotlib.ticker import FuncFormatter
    def format_rate(x, pos):
        if x >= 1e6: return f'{x/1e6:.1f}M'
        elif x >= 1e3: return f'{x/1e3:.0f}K'
        else: return f'{x:.0f}'
    
    ax.yaxis.set_major_formatter(FuncFormatter(format_rate))
    
    ax.set_title(f"Decline Curve Analysis: {UWI}\nCum Prod: {cum_prod/1e6:.2f} E6m³", fontsize=14, fontweight='bold')
    ax.set_ylabel("Gas Rate (m³/day)", fontsize=12, fontweight='bold')
    ax.set_xlabel("Date", fontsize=12, fontweight='bold')
    ax.set_ylim(bottom=0)
    
    ax.grid(True, which='major', linestyle='--', alpha=0.5)
    
    # Place Legend (Standard upper right)
    ax.legend(loc='upper right', fontsize=10, frameon=True, shadow=True)

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def exponential(t, qi, di): return qi * np.exp(-di * t)
def hyperbolic(t, qi, di, b): return qi / ((1.0 + b * di * t) ** (1.0 / b)) if b > 1e-4 else qi * np.exp(-di * t)
def harmonic(t, qi, di): return qi / (1.0 + di * t)

def get_metrics(t_current, t_abd, func, popt, cumulative_actual):
    if t_abd is None or t_abd <= t_current: return 0, cumulative_actual, 0
    t_future = np.linspace(t_current, t_abd, 1000)
    q_future = func(t_future, *popt)
    try: remaining_reserves = np.trapezoid(q_future, t_future)
    except AttributeError: remaining_reserves = np.trapezoid(q_future, t_future)
    return remaining_reserves, cumulative_actual + remaining_reserves, (t_abd - t_current) / 365.25

def find_t_abd(func, popt, q_limit, max_days_history):
    t_search = np.linspace(0, 50000, 50000)
    q_search = func(t_search, *popt)
    below_limit = np.where(q_search <= q_limit)[0]
    return t_search[below_limit[0]] if len(below_limit) > 0 else None

def dca_generator_by_well(montney_df, UWI, qabd, years):
    plot_data = montney_df[montney_df['UWI'] == UWI].copy()
    plot_data['date'] = pd.to_datetime(plot_data['date'])
    plot_data = plot_data.sort_values('date')
    
    cutoff_date = pd.Timestamp.now() - pd.DateOffset(years=years)
    plot_data = plot_data[plot_data['date'] >= cutoff_date]
    
    if plot_data.empty:
        print(f"ERROR: No data for well {UWI} in last {years} years")
        return

    if 'gas_prod_e3m3' in plot_data.columns:
        plot_data['gas_prod_m3'] = plot_data['gas_prod_e3m3'] * 1000
    elif 'gas_prod_m3' not in plot_data.columns:
        print("ERROR: Could not find 'gas_prod_e3m3' or 'gas_prod_m3' column.")
        return

    plot_data['days_online'] = (plot_data['date'] - plot_data['date'].min()).dt.days
    plot_data['days_in_period'] = plot_data['days_online'].diff().fillna(30.44)
    plot_data['rate_daily'] = plot_data['gas_prod_m3'] / plot_data['days_in_period']
    
    x_data = plot_data['days_online']
    y_data = plot_data['rate_daily']
    cum_prod = plot_data['gas_prod_m3'].sum()
    last_day = x_data.max()

    qi_guess = y_data.max()
    bounds_hyp = (0, [np.inf, np.inf, 2.0])
    
    try:
        popt_exp, _ = curve_fit(exponential, x_data, y_data, p0=[qi_guess, 0.005])
        popt_hyp, _ = curve_fit(hyperbolic, x_data, y_data, p0=[qi_guess, 0.005, 0.5], bounds=bounds_hyp)
        popt_harm, _ = curve_fit(harmonic, x_data, y_data, p0=[qi_guess, 0.005])
    except RuntimeError as e:
        print(f"Fit Failed: {e}")
        return

    t_abd_exp = find_t_abd(exponential, popt_exp, qabd, last_day)
    t_abd_hyp = find_t_abd(hyperbolic, popt_hyp, qabd, last_day)
    t_abd_harm = find_t_abd(harmonic, popt_harm, qabd, last_day)
    
    metrics_exp = get_metrics(last_day, t_abd_exp, exponential, popt_exp, cum_prod)
    metrics_hyp = get_metrics(last_day, t_abd_hyp, hyperbolic, popt_hyp, cum_prod)
    metrics_harm = get_metrics(last_day, t_abd_harm, harmonic, popt_harm, cum_prod)

    valid_abds = [t for t in [t_abd_exp, t_abd_hyp, t_abd_harm] if t is not None]
    t_max_plot = max(last_day, max(valid_abds)*1.1) if valid_abds else last_day*1.5
    t_smooth = np.linspace(x_data.min(), t_max_plot, 200)
    
    y_models = {
        'Exp': (exponential(t_smooth, *popt_exp), popt_exp, t_abd_exp, metrics_exp),
        'Hyp': (hyperbolic(t_smooth, *popt_hyp), popt_hyp, t_abd_hyp, metrics_hyp),
        'Harm': (harmonic(t_smooth, *popt_harm), popt_harm, t_abd_harm, metrics_harm)
    }
    
    date_smooth = [plot_data['date'].min() + pd.Timedelta(days=float(x)) for x in t_smooth]
    
    fig, axes = plt.subplots(3, 1, figsize=(10, 16), sharex=True)
    colors = {'Exp': '#D32F2F', 'Hyp': '#388E3C', 'Harm': '#1976D2'}
    
    for ax, (key, (y_vals, popt, t_abd, met)) in zip(axes, y_models.items()):
        ax.plot(plot_data['date'], y_data, '.', color='black', alpha=0.3, label='History')
        ax.plot(date_smooth, y_vals, color=colors[key], linewidth=2.5, label=f'{key} Fit')
        
        ax.axhline(qabd, color='orange', linestyle='--', label=f'Abd ({qabd:,.0f} m³/d)')
        if t_abd:
            abd_date = plot_data['date'].min() + pd.Timedelta(days=int(t_abd))
            ax.axvline(abd_date, color='red', linestyle=':', label=f'Abd: {abd_date.year}')
            if t_abd > last_day:
                mask = (t_smooth >= last_day) & (t_smooth <= t_abd)
                if any(mask): 
                    ax.fill_between(np.array(date_smooth)[mask], y_vals[mask], qabd, 
                                  color=colors[key], alpha=0.1, label='Remaining Reserves')

        rem, eur, life = met
        param_str = f"b={popt[2]:.2f}" if key == 'Hyp' else f"Di={popt[1]*365:.0f}%/yr"
        stats = f"EUR:  {eur/1e6:.1f} E6m³\nRem:  {rem/1e6:.1f} E6m³\nLife: {life:.1f} Yrs\n{param_str}"
        ax.text(0.98, 0.95, stats, transform=ax.transAxes, va='top', ha='right', 
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.95, edgecolor='gray'),
                fontfamily='monospace', fontsize=10)
        
        ax.set_title(f"{key} Decline", loc='left', fontweight='bold', fontsize=12)
        ax.grid(True, linestyle='--', alpha=0.4)
        ax.legend(loc='upper left', fontsize=9)
        ax.set_ylabel("Rate (m³/d)", fontweight='bold')
        ax.ticklabel_format(axis='y', style='sci', scilimits=(0,0))

    axes[-1].set_xlabel("Date", fontweight='bold')
    fig.suptitle(f"Decline Curve Analysis: {UWI}\nCumulative Production: {cum_prod/1e6:.1f} × 10⁶ m³", 
                 fontsize=14, fontweight='bold', y=0.995)
    plt.tight_layout()
    plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from matplotlib.ticker import FuncFormatter

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3
plt.rcParams['axes.titlesize'] = 12


GAS_PRICE = 3.00           # $/GJ
OPEX_PER_DAY = 450         # $/day operating cost
DRILLING_COST = 5_000_000  # $ per well
COMPLETION_COST = 3_500_000 # $ per well
DISCOUNT_RATE = 0.10       # 10% discount rate
M3_TO_GJ = 0.0373          # 1 m³ = 0.0373 GJ
ECONOMIC_LIMIT = 1000      # m³/day 


def hyperbolic(t, qi, di, b):
    """Hyperbolic decline curve model"""
    return qi / ((1.0 + b * di * t) ** (1.0 / b))

def get_eur(qi, di, b, start_day, end_day, cum_hist, econ_limit=ECONOMIC_LIMIT):
    """Calculate EUR with economic limit cutoff"""
    t_future = np.linspace(start_day, end_day, 5000)
    q_future = hyperbolic(t_future, qi, di, b)
    
    
    economic_life_days = t_future[q_future > econ_limit]
    if len(economic_life_days) > 0:
        t_future = t_future[t_future <= economic_life_days[-1]]
        q_future = q_future[:len(t_future)]
    
    rem_res = np.trapezoid(q_future, t_future)
    return cum_hist + rem_res

def calculate_npv(qi, di, b, start_day, cum_hist):
    
    years = 40
    days = np.arange(start_day, start_day + years*365, 30)  
    rates = hyperbolic(days, qi, di, b)
    
   
    rates[rates < ECONOMIC_LIMIT] = 0
    
    npv = -(DRILLING_COST + COMPLETION_COST)  
    
    for i, (day, rate) in enumerate(zip(days, rates)):
        if rate == 0:
            break
            
        month_prod = rate * 30  # m³/month
        revenue = month_prod * M3_TO_GJ * GAS_PRICE
        opex = OPEX_PER_DAY * 30
        cash_flow = revenue - opex
        
        year_frac = (day - start_day) / 365
        discount_factor = (1 + DISCOUNT_RATE) ** year_frac
        npv += cash_flow / discount_factor
    
    return npv

def calculate_payout(qi, di, b, start_day):
    """Calculate months to payout"""
    total_capex = DRILLING_COST + COMPLETION_COST
    cumulative_cf = 0
    
    for month in range(0, 120):  # Check up to 10 years
        day = start_day + month * 30
        rate = hyperbolic(np.array([day]), qi, di, b)[0]
        
        if rate < ECONOMIC_LIMIT:
            return None  
            
        month_prod = rate * 30
        revenue = month_prod * M3_TO_GJ * GAS_PRICE
        opex = OPEX_PER_DAY * 30
        cumulative_cf += (revenue - opex)
        
        if cumulative_cf >= total_capex:
            return month
    
    return None


def millions_formatter(x, pos):
    return f'{x/1e6:.1f}'

def rate_formatter(x, pos):
    if x >= 1000:
        return f'{x/1e3:.0f}k'
    return f'{x:.0f}'

def currency_formatter(x, pos):
    if abs(x) >= 1e6:
        return f'${x/1e6:.1f}M'
    elif abs(x) >= 1e3:
        return f'${x/1e3:.0f}k'
    return f'${x:.0f}'


def probabilistic_dashboard_with_economics(UWI, montney_df, iterations=2000):
    
    # --- A. Data Prep ---
    plot_data = montney_df[montney_df['UWI'] == UWI].copy()
    plot_data['date'] = pd.to_datetime(plot_data['date'])
    plot_data = plot_data.sort_values('date')
    
    if plot_data.empty: 
        print(f"No data found for {UWI}")
        return
    
    plot_data['days_online'] = (plot_data['date'] - plot_data['date'].min()).dt.days
    plot_data['days_in_month'] = plot_data['days_online'].diff().fillna(30.44)
    
    x_data = plot_data['days_online']
    y_data = (plot_data['gas_prod_e3m3'] * 1000) / plot_data['days_in_month']
    cum_prod = (plot_data['gas_prod_e3m3'] * 1000).sum()
    
    current_day = x_data.max()
    t_end = 15000  # 40 years
    
    # --- B. Fitting ---
    qi_guess = y_data.max()
    bounds = (0, [np.inf, np.inf, 2.0])
    
    try:
        p_opt, p_cov = curve_fit(hyperbolic, x_data, y_data, p0=[qi_guess, 0.005, 0.5], bounds=bounds)
    except:
        print(f"Fit failed for {UWI}")
        return
    
    # --- C. Monte Carlo Simulation ---
    samples = np.random.multivariate_normal(p_opt, p_cov, size=iterations)
    valid_samples = []
    for sample in samples:
        qi_s, di_s, b_s = sample
        if qi_s > 100 and di_s > 0.00001 and 0 < b_s < 2.5:
            valid_samples.append(sample)
    
    valid_samples = np.array(valid_samples)
    print(f"✓ Valid samples: {len(valid_samples)}/{iterations} ({len(valid_samples)/iterations*100:.1f}%)")
    
    # --- D. Calculate EURs, NPVs, IRRs ---
    simulated_eurs = []
    simulated_npvs = []
    simulated_payouts = []
    
    for (qi_s, di_s, b_s) in valid_samples:
        eur = get_eur(qi_s, di_s, b_s, current_day, t_end, cum_prod)
        npv = calculate_npv(qi_s, di_s, b_s, current_day, cum_hist=cum_prod)
        payout = calculate_payout(qi_s, di_s, b_s, current_day)
        
        simulated_eurs.append(eur)
        simulated_npvs.append(npv)
        simulated_payouts.append(payout if payout else 999) 
    
    simulated_eurs = np.array(simulated_eurs)
    simulated_npvs = np.array(simulated_npvs)
    simulated_payouts = np.array(simulated_payouts)
    
    # --- E. Find Percentiles ---
    p90_eur = np.percentile(simulated_eurs, 10)
    p50_eur = np.percentile(simulated_eurs, 50)
    p10_eur = np.percentile(simulated_eurs, 90)
    
    p90_npv = np.percentile(simulated_npvs, 10)
    p50_npv = np.percentile(simulated_npvs, 50)
    p10_npv = np.percentile(simulated_npvs, 90)
    
    # Economic success rate (NPV > 0)
    success_rate = (simulated_npvs > 0).sum() / len(simulated_npvs) * 100
    
    # Find parameters for each case
    idx_p90 = (np.abs(simulated_eurs - p90_eur)).argmin()
    idx_p50 = (np.abs(simulated_eurs - p50_eur)).argmin()
    idx_p10 = (np.abs(simulated_eurs - p10_eur)).argmin()
    
    params_p90 = valid_samples[idx_p90]
    params_p50 = valid_samples[idx_p50]
    params_p10 = valid_samples[idx_p10]
    
    payout_p90 = simulated_payouts[idx_p90]
    payout_p50 = simulated_payouts[idx_p50]
    payout_p10 = simulated_payouts[idx_p10]
    
    # --- F. PRINT BUSINESS SUMMARY ---
    print("\n" + "="*70)
    print(f"ECONOMIC SUMMARY: {UWI}")
    print("="*70)
    print(f"\nRESERVES (EUR):")
    print(f"   P90 (Proved):    {p90_eur/1e6:>8.2f} × 10⁶ m³  ({p90_eur*M3_TO_GJ/1e9:.3f} Bcf)")
    print(f"   P50 (Best Est):  {p50_eur/1e6:>8.2f} × 10⁶ m³  ({p50_eur*M3_TO_GJ/1e9:.3f} Bcf)")
    print(f"   P10 (Upside):    {p10_eur/1e6:>8.2f} × 10⁶ m³  ({p10_eur*M3_TO_GJ/1e9:.3f} Bcf)")
    
    print(f"\nNET PRESENT VALUE (@ {DISCOUNT_RATE*100:.0f}% discount, ${GAS_PRICE}/GJ):")
    print(f"   P90 (Proved):    ${p90_npv/1e6:>6.2f}M")
    print(f"   P50 (Best Est):  ${p50_npv/1e6:>6.2f}M")
    print(f"   P10 (Upside):    ${p10_npv/1e6:>6.2f}M")
    
    print(f"\nPAYOUT PERIOD:")
    print(f"   P90 (Proved):    {payout_p90:>3.0f} months" if payout_p90 < 999 else "   P90 (Proved):    Never")
    print(f"   P50 (Best Est):  {payout_p50:>3.0f} months" if payout_p50 < 999 else "   P50 (Best Est):  Never")
    print(f"   P10 (Upside):    {payout_p10:>3.0f} months" if payout_p10 < 999 else "   P10 (Upside):    Never")
    
    print(f"\nECONOMIC SUCCESS RATE: {success_rate:.1f}% of wells are NPV positive")
    
    print(f"\nDECLINE PARAMETERS:")
    print(f"   P90: qi={params_p90[0]/1000:.1f}k m³/d, Di={params_p90[1]:.5f}, b={params_p90[2]:.3f}")
    print(f"   P50: qi={params_p50[0]/1000:.1f}k m³/d, Di={params_p50[1]:.5f}, b={params_p50[2]:.3f}")
    print(f"   P10: qi={params_p10[0]/1000:.1f}k m³/d, Di={params_p10[1]:.5f}, b={params_p10[2]:.3f}")
    
    print(f"\nINVESTMENT RECOMMENDATION:")
    if success_rate >= 80:
        print(f"   STRONG BUY - {success_rate:.0f}% success rate, P50 NPV of ${p50_npv/1e6:.1f}M")
    elif success_rate >= 60:
        print(f"   CONDITIONAL - {success_rate:.0f}% success rate. Acceptable at current prices.")
    else:
        print(f"   MARGINAL - Only {success_rate:.0f}% success rate. High risk.")
    
    print("="*70 + "\n")
    
    # --- G. PLOTTING ---
    fig = plt.figure(figsize=(18, 12))
    gs = fig.add_gridspec(3, 2, height_ratios=[2, 1.5, 1.5])
    
    ax_time = fig.add_subplot(gs[0, :])
    ax_cum  = fig.add_subplot(gs[1, 0])
    ax_hist_eur = fig.add_subplot(gs[1, 1])
    ax_hist_npv = fig.add_subplot(gs[2, 0])
    ax_econ_table = fig.add_subplot(gs[2, 1])
    ax_econ_table.axis('off')
    
    # --- GRAPH 1: Rate vs Time ---
    t_smooth = np.linspace(0, t_end, 1000)
    date_smooth = [plot_data['date'].min() + pd.Timedelta(days=float(t)) for t in t_smooth]
    
    y_p10 = hyperbolic(t_smooth, *params_p10)
    y_p50 = hyperbolic(t_smooth, *params_p50)
    y_p90 = hyperbolic(t_smooth, *params_p90)
    
    ax_time.plot(plot_data['date'], y_data, '.', color='#555555', alpha=0.4, markersize=5, label='Historical Production')
    ax_time.plot(date_smooth, y_p10, color='#2ca02c', ls='--', lw=2, label=f'P10 (Upside) - {p10_eur/1e6:.1f}M m³')
    ax_time.plot(date_smooth, y_p50, color='#1f77b4', lw=3, label=f'P50 (Base Case) - {p50_eur/1e6:.1f}M m³')
    ax_time.plot(date_smooth, y_p90, color='#d62728', ls='--', lw=2, label=f'P90 (Proved) - {p90_eur/1e6:.1f}M m³')
    ax_time.fill_between(date_smooth, y_p90, y_p10, color='gray', alpha=0.15, label='80% Confidence Band')
    ax_time.axhline(ECONOMIC_LIMIT, color='red', ls=':', lw=1.5, alpha=0.5, label=f'Economic Limit ({ECONOMIC_LIMIT} m³/d)')
    
    ax_time.set_title(f"Production Forecast - {UWI}", fontweight='bold', fontsize=14)
    ax_time.set_ylabel("Gas Rate (m³/d)", fontsize=12, fontweight='bold')
    ax_time.yaxis.set_major_formatter(FuncFormatter(rate_formatter))
    ax_time.legend(loc='upper right', frameon=True, fontsize=10)
    ax_time.set_ylim(bottom=0)
    ax_time.set_xlabel("Date", fontsize=11)
    
    # --- GRAPH 2: Rate vs Cumulative ---
    def get_cum_curve(params, t_arr):
        q = hyperbolic(t_arr, *params)
        dt = t_arr[1] - t_arr[0]
        return np.cumsum(q) * dt
    
    cum_p10 = get_cum_curve(params_p10, t_smooth)
    cum_p50 = get_cum_curve(params_p50, t_smooth)
    cum_p90 = get_cum_curve(params_p90, t_smooth)
    
    ax_cum.plot(cum_p10, y_p10, color='#2ca02c', ls='--', lw=2, alpha=0.7, label='P10')
    ax_cum.plot(cum_p50, y_p50, color='#1f77b4', lw=2.5, label='P50')
    ax_cum.plot(cum_p90, y_p90, color='#d62728', ls='--', lw=2, alpha=0.7, label='P90')
    
    cum_history = (plot_data['gas_prod_e3m3'] * 1000).cumsum()
    ax_cum.plot(cum_history, y_data, 'o', color='#555555', alpha=0.3, markersize=3, label='History')
    
    ax_cum.set_title("Diagnostic Plot: Rate vs. Cumulative", fontweight='bold')
    ax_cum.set_xlabel("Cumulative Production (10⁶ m³)", fontsize=11)
    ax_cum.set_ylabel("Gas Rate (m³/d)", fontsize=11)
    ax_cum.xaxis.set_major_formatter(FuncFormatter(millions_formatter))
    ax_cum.yaxis.set_major_formatter(FuncFormatter(rate_formatter))
    ax_cum.legend(loc='upper right', fontsize=9)
    ax_cum.grid(True, which="both", ls="--", alpha=0.4)
    
    # --- GRAPH 3: EUR Distribution ---
    eur_millions = simulated_eurs / 1e6
    
    counts, bins, patches = ax_hist_eur.hist(eur_millions, bins=40, color='#1f77b4', alpha=0.3, edgecolor='white')
    
    ax_hist_eur.axvline(p90_eur/1e6, color='#d62728', lw=2.5, label=f'P90: {p90_eur/1e6:.1f}M m³')
    ax_hist_eur.axvline(p50_eur/1e6, color='#1f77b4', lw=2.5, label=f'P50: {p50_eur/1e6:.1f}M m³')
    ax_hist_eur.axvline(p10_eur/1e6, color='#2ca02c', lw=2.5, label=f'P10: {p10_eur/1e6:.1f}M m³')
    
    ax_hist_eur.set_title("EUR Probability Distribution", fontweight='bold')
    ax_hist_eur.set_xlabel("EUR (10⁶ m³)", fontsize=11)
    ax_hist_eur.set_ylabel("Frequency", fontsize=11)
    ax_hist_eur.legend(fontsize=9)
    
    # --- GRAPH 4: NPV Distribution ---
    npv_millions = simulated_npvs / 1e6
    
    counts, bins, patches = ax_hist_npv.hist(npv_millions, bins=40, color='#2ca02c', alpha=0.3, edgecolor='white')
    
    
    for i, patch in enumerate(patches):
        if bins[i] < 0:
            patch.set_facecolor('#d62728')
            patch.set_alpha(0.4)
    
    ax_hist_npv.axvline(0, color='black', lw=2, ls='--', alpha=0.7, label='Breakeven')
    ax_hist_npv.axvline(p90_npv/1e6, color='#d62728', lw=2.5, label=f'P90: ${p90_npv/1e6:.1f}M')
    ax_hist_npv.axvline(p50_npv/1e6, color='#1f77b4', lw=2.5, label=f'P50: ${p50_npv/1e6:.1f}M')
    ax_hist_npv.axvline(p10_npv/1e6, color='#2ca02c', lw=2.5, label=f'P10: ${p10_npv/1e6:.1f}M')
    
    ax_hist_npv.set_title(f"NPV Distribution (${GAS_PRICE}/GJ, {DISCOUNT_RATE*100:.0f}% Discount)", fontweight='bold')
    ax_hist_npv.set_xlabel("NPV ($M)", fontsize=11)
    ax_hist_npv.set_ylabel("Frequency", fontsize=11)
    ax_hist_npv.legend(fontsize=9)
    ax_hist_npv.xaxis.set_major_formatter(FuncFormatter(lambda x, p: f'${x:.1f}M'))
    
    # --- GRAPH 5: Economics Table ---
    table_data = [
        ['Metric', 'P90 (Proved)', 'P50 (Base)', 'P10 (Upside)'],
        ['EUR (10⁶ m³)', f'{p90_eur/1e6:.2f}', f'{p50_eur/1e6:.2f}', f'{p10_eur/1e6:.2f}'],
        ['EUR (Bcf)', f'{p90_eur*M3_TO_GJ/1e9:.3f}', f'{p50_eur*M3_TO_GJ/1e9:.3f}', f'{p10_eur*M3_TO_GJ/1e9:.3f}'],
        ['NPV ($M)', f'{p90_npv/1e6:.2f}', f'{p50_npv/1e6:.2f}', f'{p10_npv/1e6:.2f}'],
        ['Payout (mo)', f'{payout_p90:.0f}' if payout_p90<999 else 'Never', 
                        f'{payout_p50:.0f}' if payout_p50<999 else 'Never',
                        f'{payout_p10:.0f}' if payout_p10<999 else 'Never'],
        ['qi (k m³/d)', f'{params_p90[0]/1000:.1f}', f'{params_p50[0]/1000:.1f}', f'{params_p10[0]/1000:.1f}'],
        ['Di (/day)', f'{params_p90[1]:.5f}', f'{params_p50[1]:.5f}', f'{params_p10[1]:.5f}'],
        ['b-factor', f'{params_p90[2]:.3f}', f'{params_p50[2]:.3f}', f'{params_p10[2]:.3f}'],
    ]
    
    table = ax_econ_table.table(cellText=table_data, cellLoc='center', loc='center',
                                  colWidths=[0.35, 0.22, 0.22, 0.22])
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 2.5)
    
    # Style header row
    for i in range(4):
        table[(0, i)].set_facecolor('#003f5c')
        table[(0, i)].set_text_props(weight='bold', color='white')
    
    # Style metric column
    for i in range(1, len(table_data)):
        table[(i, 0)].set_facecolor('#f0f0f0')
        table[(i, 0)].set_text_props(weight='bold')
    
    ax_econ_table.set_title(f"Economic Summary\n(${GAS_PRICE}/GJ gas, ${OPEX_PER_DAY}/d OPEX, {DISCOUNT_RATE*100:.0f}% discount)\n Success Rate: {success_rate:.1f}%", 
                             fontweight='bold', pad=20, fontsize=11)
    
    # --- FOOTER ---
    assumptions_text = (f"Assumptions: Drilling=${DRILLING_COST/1e6:.1f}M | Completion=${COMPLETION_COST/1e6:.1f}M | "
                       f"OPEX=${OPEX_PER_DAY}/d | Econ Limit={ECONOMIC_LIMIT} m³/d | {iterations} Monte Carlo iterations")
    
    plt.figtext(0.02, 0.01, assumptions_text, 
                horizontalalignment='left', fontsize=9, style='italic', color='#666666')
    
    plt.figtext(0.98, 0.01, f"Data: BC Energy Regulator | Analysis: Shahbaz S. Mann", 
                horizontalalignment='right', fontsize=9, style='italic', color='#666666')
    

    fig.suptitle(f"Probabilistic Economic Forecast: {UWI}", fontsize=16, fontweight='bold', y=0.98)
    plt.tight_layout(rect=[0, 0.02, 1, 0.97])
    plt.show()
    
    return {
        'eur_p90': p90_eur, 'eur_p50': p50_eur, 'eur_p10': p10_eur,
        'npv_p90': p90_npv, 'npv_p50': p50_npv, 'npv_p10': p10_npv,
        'success_rate': success_rate,
        'params_p90': params_p90, 'params_p50': params_p50, 'params_p10': params_p10,
        'simulated_eurs': simulated_eurs, 'simulated_npvs': simulated_npvs
    }



results = probabilistic_dashboard_with_economics('202C065C094H0400', montney_df, iterations=2000)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def plot_type_curve_professional(df, pool_id='A', start_year=2020, end_year=2022, author_name="Your Name"):
    # --- 1. DATA PREP ---
    df = df.copy()
    df['date'] = pd.to_datetime(df['date'])
    
    # Identify Start Date for every well
    df['start_date'] = df.groupby('UWI')['date'].transform('min')
    
    # Filter for specific Pool and Vintage
    vintage_wells = df[
        (df['pool_seq'] == pool_id) & 
        (df['start_date'].dt.year >= start_year) &
        (df['start_date'].dt.year <= end_year)
    ].copy()
    
    if vintage_wells.empty:
        print(f"No wells found for Pool {pool_id} in {start_year}-{end_year}")
        return

    # Calculate Normalized Time (Months on Production)
    vintage_wells['month_num'] = (
        (vintage_wells['date'].dt.year - vintage_wells['start_date'].dt.year) * 12 + 
        (vintage_wells['date'].dt.month - vintage_wells['start_date'].dt.month)
    )
    
    # --- 2. AGGREGATE STATISTICS ---
  
    stats = vintage_wells.groupby('month_num')['gas_prod_e3m3'].agg(
        mean='mean',
        p50='median',
        p10=lambda x: x.quantile(0.90), # P10 (Upside)
        p90=lambda x: x.quantile(0.10), # P90 (Downside)
        count='count'
    )
    
    # Cut off tail (<10 wells)
    stats = stats[stats['count'] >= 10]
    
    # --- 3. PLOTTING ---
    fig, ax = plt.subplots(figsize=(12, 7), facecolor='white')
    
    # A. Plot The Cone (P10-P90)
    ax.fill_between(stats.index, stats['p90'], stats['p10'], 
                    color='#b0b0b0', alpha=0.15, label='P10-P90 Range')
    
    # B. Plot The Curves
    ax.plot(stats.index, stats['mean'], 
            color='#003f5c', linewidth=3, label='Mean Average')
    
    ax.plot(stats.index, stats['p50'], 
            color='#ffa600', linewidth=2, linestyle='--', label='P50 (Median)')

    # C. Plot Well Count (Secondary Axis)
    ax_count = ax.twinx()
    ax_count.bar(stats.index, stats['count'], color='gray', alpha=0.1, width=1.0)
    ax_count.set_ylabel("Well Count (n)", color='gray', fontsize=9)
    ax_count.grid(False)

    # --- 4. STYLING ---
    ax.set_title(f"Montney Pool '{pool_id}' Type Curve ({start_year}-{end_year} Vintage)", 
                 fontsize=14, fontweight='bold', pad=15)
    
   
    ax.set_ylabel("Monthly Gas Production ($10^3 m^3$)", fontweight='bold')
    
    # Just format with comma
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{x:,.0f}'))
    
    ax.set_xlabel("Months on Production", fontweight='bold')
    ax.set_xlim(0, stats.index.max())
    ax.set_ylim(bottom=0)
    ax.grid(True, which='major', alpha=0.3)
    ax.legend(loc='upper right', frameon=True)
    
    # Peak Rate Annotation
    peak_rate = stats['mean'].max()
    peak_month = stats['mean'].idxmax()
    
    ax.annotate(f"Peak Rate: {peak_rate:,.0f} x $10^3 m^3$", 
                xy=(peak_month, peak_rate), xytext=(peak_month+5, peak_rate),
                arrowprops=dict(facecolor='black', shrink=0.05))

    # Footer
    plt.figtext(0.99, 0.01, f"Source: BC Energy Regulator | Analysis & Viz: {author_name}", 
                horizontalalignment='right', fontsize=10, style='italic', color='#555555',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=5))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.1)
    plt.show()

plot_type_curve_professional(montney_df, pool_id='A', start_year=2022, end_year=2024, author_name="Shahbaz S.Mann")
plot_type_curve_professional(montney_df, pool_id='A', start_year=2012, end_year=2014, author_name="Shahbaz S.Mann")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy import stats

def plot_prob_plot_professional(df, pool_id='A', start_year=2020, end_year=2022, author_name="Your Name"):
    # --- 1. DATA PREP ---
    df = df.copy()
    df['date'] = pd.to_datetime(df['date'])
    
    # Identify vintage
    df['start_date'] = df.groupby('UWI')['date'].transform('min')
    
    # Filter for target wells
    vintage_wells = df[
        (df['pool_seq'] == pool_id) & 
        (df['start_date'].dt.year >= start_year) &
        (df['start_date'].dt.year <= end_year)
    ].copy()
    
    if vintage_wells.empty:
        print(f"No wells found for Pool {pool_id}")
        return

   
    well_values = vintage_wells.groupby('UWI')['gas_prod_e3m3'].sum().sort_values(ascending=False)
    
   #ghost well  filtration
    well_values = well_values[well_values > 1000]
    
 
    values = well_values.values
    n = len(values)
    

    ranks = np.arange(1, n + 1)
    probabilities = (ranks / (n + 1)) * 100  # Convert to %
    
 
    fig, ax = plt.subplots(figsize=(10, 7), facecolor='white')
    

    ax.scatter(probabilities, values, color='#003f5c', s=30, alpha=0.6, label='Individual Wells')
    
   
    try:
        shape, loc, scale = stats.lognorm.fit(values)
        x_fit = np.linspace(1, 99, 100)
      
        y_fit = stats.lognorm.isf(x_fit/100, shape, loc, scale)
        ax.plot(x_fit, y_fit, color='#ffa600', linestyle='--', linewidth=2, label='Log-Normal Trend')
    except:
        pass # Skip if fit fails

 
    breakeven_vol = 50000 
    
    ax.axhline(breakeven_vol, color='#bc5090', linestyle='-', linewidth=2, label=f'Breakeven ({breakeven_vol:,.0f})')
    
 
    ax.fill_between(probabilities, breakeven_vol, values.max()*1.1, 
                    where=(values >= breakeven_vol), 
                    color='#b0b0b0', alpha=0.1, transform=ax.get_xaxis_transform())

   
    ax.set_title(f"Cumulative Probability: Pool '{pool_id}' ({start_year}-{end_year})", 
                 fontsize=14, fontweight='bold', pad=15)
    
    ax.set_xlabel("Cumulative Probability (P-Value) %", fontweight='bold')
    ax.set_ylabel("Cumulative Gas Production ($10^3 m^3$)", fontweight='bold')
    

    ax.set_yscale('log')
    
  
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{x:,.0f}'))
    
   
    ax.set_xlim(1, 99)
   
    ax.set_xticks([10, 50, 90])
    ax.set_xticklabels(['P10\n(Upside)', 'P50\n(Average)', 'P90\n(Risk)'])
    
    ax.grid(True, which="both", alpha=0.3)
    ax.legend()


    plt.figtext(0.99, 0.99, f"Source: BC Energy Regulator | Analysis & Viz: {author_name}", 
                horizontalalignment='right', fontsize=10, style='italic', color='#555555',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=5))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.1)
    plt.show()


plot_prob_plot_professional(montney_df, pool_id='A', start_year=2020, end_year=2022, author_name="Shahbaz S.Mann")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

def plot_wgr_type_curve(df, pool_id='A', start_year=2020, end_year=2022, author_name="Shahbaz S.Mann"):
    # --- 1. DATA PREP ---
    df = df.copy()
    df['date'] = pd.to_datetime(df['date'])
    
    if 'water_prod_m3' not in df.columns:
        print("⚠️ Error: 'water_prod_m3' column not found.")
        return

    df['start_date'] = df.groupby('UWI')['date'].transform('min')
    
    vintage_wells = df[
        (df['pool_seq'] == pool_id) & 
        (df['start_date'].dt.year >= start_year) &
        (df['start_date'].dt.year <= end_year)
    ].copy()

    if vintage_wells.empty:
        print(f"No wells found for Pool {pool_id}")
        return

    vintage_wells['month_num'] = (
        (vintage_wells['date'].dt.year - vintage_wells['start_date'].dt.year) * 12 + 
        (vintage_wells['date'].dt.month - vintage_wells['start_date'].dt.month)
    )
    
    # --- 2. CALCULATE WGR ---
    vintage_wells = vintage_wells[vintage_wells['gas_prod_e3m3'] > 0]
    vintage_wells['wgr'] = vintage_wells['water_prod_m3'] / vintage_wells['gas_prod_e3m3']
    vintage_wells = vintage_wells[vintage_wells['wgr'] < 50]

    # --- 3. AGGREGATE ---
    stats = vintage_wells.groupby('month_num')['wgr'].agg(
        mean='mean',
        p50='median',
        p10=lambda x: x.quantile(0.90), 
        p90=lambda x: x.quantile(0.10), 
        count='count'
    )
    
    stats = stats[stats['count'] >= 10]

    # --- 4. PLOTTING ---
    fig, ax = plt.subplots(figsize=(12, 7), facecolor='white')
    
    # Plot Cone & Curves
    ax.fill_between(stats.index, stats['p90'], stats['p10'], 
                    color='#003f5c', alpha=0.1, label='P10-P90 Range')
    
    ax.plot(stats.index, stats['mean'], 
            color='#003f5c', linewidth=3, label='Mean WGR')
    
    ax.plot(stats.index, stats['p50'], 
            color='#ffa600', linewidth=2, linestyle='--', label='P50 (Median)')
    
    # Styling
    ax.set_title(f"Water-Gas Ratio (WGR) Evolution: Pool '{pool_id}'", 
                 fontsize=14, fontweight='bold', pad=15)
    
    ax.set_ylabel("WGR ($m^3$ Water / $10^3 m^3$ Gas)", fontweight='bold')
    ax.set_xlabel("Months on Production", fontweight='bold')
    
    ax.set_xlim(0, stats.index.max())
    ax.set_ylim(bottom=0)
    
    ax.grid(True, which='major', alpha=0.3)
    

    ax.legend(loc='upper left', frameon=True)
    

    start_wgr = stats['mean'].iloc[0]
    end_wgr = stats['mean'].iloc[-1]
    trend_color = 'red' if end_wgr > start_wgr * 1.5 else 'green'
    trend_text = "Rising Water Cut (Risk)" if end_wgr > start_wgr * 1.5 else "Stable Water Profile"
    
  
    ax.text(0.98, 0.95, trend_text, transform=ax.transAxes, 
            color=trend_color, fontweight='bold', ha='right', va='top',
            bbox=dict(facecolor='white', edgecolor=trend_color, alpha=0.9))


    plt.figtext(0.99, 0.01, f"Source: BC Energy Regulator | Analysis & Viz: {author_name}", 
                horizontalalignment='right', fontsize=10, style='italic', color='#555555',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=5))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.1)
    plt.show()


plot_wgr_type_curve(montney_df, pool_id='A', start_year=2020, end_year=2022, author_name="Shahbaz S.Mann")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D

def plot_plateau_analysis(df, pool_id='A', start_year=2018, end_year=2022, author_name="Your Name"):
    # --- 1. DATA PREP ---
    df = df.copy()
    df['date'] = pd.to_datetime(df['date'])
    
   
    if 'gas_prod_e3m3' not in df.columns:
         if 'gas_prod_m3' in df.columns:
             df['gas_prod_e3m3'] = df['gas_prod_m3'] / 1000
         else:
             print("⚠️ Error: Gas production column not found.")
             return

  
    df['start_date'] = df.groupby('UWI')['date'].transform('min')
    df['spud_year'] = df['start_date'].dt.year
    
    target_wells = df[
        (df['pool_seq'] == pool_id) & 
        (df['spud_year'] >= start_year) & 
        (df['spud_year'] <= end_year)
    ].copy()

    if target_wells.empty:
        print(f"No wells found for Pool {pool_id} in {start_year}-{end_year}")
        return

   
    def get_plateau_metrics(group):
        peak_rate = group['gas_prod_e3m3'].max()
        if peak_rate == 0: return pd.Series([0, 0], index=['peak_rate', 'plateau_months'])
        
        plateau_threshold = peak_rate * 0.90
        plateau_months = group[group['gas_prod_e3m3'] >= plateau_threshold].shape[0]
        
        return pd.Series([peak_rate, plateau_months], index=['peak_rate', 'plateau_months'])

    well_metrics = target_wells.groupby('UWI').apply(get_plateau_metrics).reset_index()

    year_map = target_wells[['UWI', 'spud_year']].drop_duplicates()
    well_metrics = well_metrics.merge(year_map, on='UWI')
    
    well_metrics = well_metrics[well_metrics['peak_rate'] > 500]


    fig, ax = plt.subplots(figsize=(14, 9), facecolor='white')
    

    max_rate = well_metrics['peak_rate'].max()
    mean_rate = well_metrics['peak_rate'].mean()
    
 
    zone1 = plt.Rectangle((0, max_rate*0.7), 5, max_rate*0.3, 
                          color='#e63946', alpha=0.12, zorder=1)
    ax.add_patch(zone1)
    

    zone2 = plt.Rectangle((10, 0), 25, mean_rate*0.8, 
                          color='#06a77d', alpha=0.12, zorder=1)
    ax.add_patch(zone2)
    
   
    import numpy as np
    np.random.seed(42)
    jitter_x = np.random.normal(0, 0.15, size=len(well_metrics))
    jitter_y = np.random.normal(0, max_rate*0.005, size=len(well_metrics))
    
    scatter = ax.scatter(
        well_metrics['plateau_months'] + jitter_x, 
        well_metrics['peak_rate'] + jitter_y, 
        c=well_metrics['spud_year'], 
        cmap='plasma', 
        s=60,  
        alpha=0.65,          
        edgecolors='white', 
        linewidth=0.8,
        zorder=3
    )

   
    ax.set_title(
        f"Well Performance Diagnostics: Peak Rate vs. Plateau Duration\nPool '{pool_id}' ({start_year}-{end_year})", 
        fontsize=15, fontweight='bold', pad=20
    )
    
    ax.set_xlabel("Plateau Duration (Months at >90% of Peak Rate)", fontweight='bold', fontsize=12)
    ax.set_ylabel("Peak Monthly Production Rate (10³ m³/month)", fontweight='bold', fontsize=12)
    
   
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{x:,.0f}'))
    ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
    
    
    ax.grid(True, which='major', alpha=0.25, linestyle='--', linewidth=0.8, zorder=0)
    ax.set_axisbelow(True)
    
    
    zone_legend_elements = [
        mpatches.Patch(facecolor='#e63946', alpha=0.3, edgecolor='#e63946', linewidth=1.5,
                      label='Aggressive Drawdown\n(High IP, Short Plateau)'),
        mpatches.Patch(facecolor='#06a77d', alpha=0.3, edgecolor='#06a77d', linewidth=1.5,
                      label='Managed/Choked Flow\n(Sustained Production)')
    ]
    
    legend1 = ax.legend(handles=zone_legend_elements, loc='upper right', 
                       framealpha=0.95, edgecolor='gray', title='Production Zones',
                       title_fontsize=11, fontsize=10)
    ax.add_artist(legend1) 
    
  
    cbar = plt.colorbar(scatter, ax=ax, pad=0.02)
    cbar.set_label('Spud Year', fontweight='bold', fontsize=11)
    cbar.locator = ticker.MaxNLocator(integer=True)
    cbar.update_ticks()

    
    stats_text = (
        f"n = {len(well_metrics)} wells\n"
        f"Median Peak: {well_metrics['peak_rate'].median():,.0f} 10³m³\n"
        f"Median Plateau: {well_metrics['plateau_months'].median():.1f} months"
    )
    ax.text(0.02, 0.98, stats_text, transform=ax.transAxes,
           fontsize=10, verticalalignment='top',
           bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

   
    plt.figtext(0.99, 0.01, f"Source: BC Energy Regulator | Analysis: {author_name}", 
                horizontalalignment='right', fontsize=9, style='italic', color='#555555',
                bbox=dict(facecolor='white', alpha=0.9, edgecolor='lightgray', pad=5))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.08)
    plt.show()


plot_plateau_analysis(montney_df, pool_id='A', start_year=2018, end_year=2022, author_name="Shahbaz S.Mann")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

def plot_montney_grid_map(df, pool_id='A', author_name="Your Name"):

    df = df.copy()
    df['date'] = pd.to_datetime(df['date'])
    

    df['start_date'] = df.groupby('UWI')['date'].transform('min')
    
 
    pool_wells = df[df['pool_seq'] == pool_id].copy()
    
    if pool_wells.empty:
        print(f"No wells found for Pool {pool_id}")
        return


    def parse_location(uwi):
        try:
            clean_uwi = str(uwi).replace('/', '').replace('-', '').replace('.', '')
            if len(clean_uwi) >= 12:
                w_index = clean_uwi.find('W')
                if w_index > 4:
                    rng = int(clean_uwi[w_index-2 : w_index]) 
                    twp = int(clean_uwi[w_index-5 : w_index-2])
                    return pd.Series([twp, rng])
            return pd.Series([None, None])
        except:
            return pd.Series([None, None])

    pool_wells[['Twp', 'Rge']] = pool_wells['UWI'].apply(parse_location)
    pool_wells = pool_wells.dropna(subset=['Twp', 'Rge'])
    
    
    if 'gas_prod_e3m3' in pool_wells.columns:
        bubble_size = pool_wells.groupby('UWI')['gas_prod_e3m3'].sum()
    else:
        
        bubble_size = pool_wells.groupby('UWI')['gas_prod_m3'].sum() / 1000
        
   
    map_data = pool_wells[['UWI', 'Twp', 'Rge', 'start_date']].drop_duplicates(subset=['UWI']).copy()
    map_data = map_data.set_index('UWI')
    map_data['cum_gas'] = bubble_size
    map_data['spud_year'] = map_data['start_date'].dt.year

   
    fig, ax = plt.subplots(figsize=(12, 10), facecolor='white')
    

    scatter = ax.scatter(
        map_data['Rge'], 
        map_data['Twp'], 
        s=map_data['cum_gas'] / 600,  
        c=map_data['spud_year'],     
        cmap='viridis',              
        alpha=0.6,                   
        edgecolors='black', 
        linewidth=0.5
    )
    
    
    ax.set_title(f"Reservoir Grid Map: Montney Pool '{pool_id}'\n(Bubble Size = Reserves, Color = Vintage)", 
                 fontsize=14, fontweight='bold', pad=20)
    
    ax.set_xlabel("Range (West)", fontweight='bold')
    ax.set_ylabel("Township (North)", fontweight='bold')
    
 
    ax.invert_xaxis()
    
  
    ax.grid(True, which='both', linestyle='--', alpha=0.4)
    ax.minorticks_on()
    

    cbar = plt.colorbar(scatter)
    cbar.set_label("Spud Year", fontweight='bold')
  
    cbar.locator = ticker.MaxNLocator(integer=True)
    cbar.update_ticks()

  
    best_twp = map_data.groupby('Twp')['cum_gas'].sum().idxmax()
    ax.axhline(best_twp, color='red', linestyle='--', alpha=0.3)
  

 
    plt.figtext(0.99, 0.01, f"Source: BC Energy Regulator | Analysis & Viz: {author_name}", 
                horizontalalignment='right', fontsize=10, style='italic', color='#555555',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=5))

    plt.tight_layout()
    plt.show()

# Run it
plot_montney_grid_map(montney_df, pool_id='A', author_name="Shahbaz S.Mann")
```
